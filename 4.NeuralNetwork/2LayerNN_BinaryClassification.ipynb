```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import multivariate_normal, permutation
import pandas as pd
from pandas import DataFrame, Series

#np.random.seed(20160614)
#tf.set_random_seed(20160614)

np.random.seed(20180427)
tf.set_random_seed(20180427)
```

```python
#mu , variance , n , t 
data0 = multivariate_normal([-7,-7],np.eye(2)*18,30)
df0 = DataFrame(data0, columns=['x1','x2'])
df0['t'] = 1

data1 = multivariate_normal([-7,7],np.eye(2)*18,30)
df1 = DataFrame(data1, columns=['x1','x2'])
df1['t'] = 0

data2 = multivariate_normal([7,-7],np.eye(2)*18,30)
df2 = DataFrame(data2, columns=['x1','x2'])
df2['t'] = 0

data3 = multivariate_normal([7,7],np.eye(2)*18,30)
df3 = DataFrame(data3, columns=['x1','x2'])
df3['t'] = 1

#permutation() : 순서 무작위로 바꿈 셔플 
df = pd.concat([df0, df1, df2, df3], ignore_index=True)
train_set = df.reindex(permutation(df.index)).reset_index(drop=True)

train_set
```

```python
train_x = train_set[['x1','x2']].as_matrix()
train_t = train_set['t'].as_matrix().reshape([len(train_set), 1])
```

```python
nodeNum1 = 2
nodeNum2 = 2

x = tf.placeholder(tf.float32, [None, 2])

w1 = tf.Variable(tf.truncated_normal([2, nodeNum1]))
b1 = tf.Variable(tf.zeros([nodeNum1]))
hidden1 = tf.nn.tanh(tf.matmul(x, w1) + b1)

w2 = tf.Variable(tf.truncated_normal([nodeNum1, nodeNum2]))
b2 = tf.Variable(tf.zeros([nodeNum2]))
hidden2 = tf.nn.tanh(tf.matmul(hidden1, w2) + b2)

w0 = tf.Variable(tf.zeros([nodeNum2, 1]))
b0 = tf.Variable(tf.zeros([1]))
p = tf.nn.sigmoid(tf.matmul(hidden2, w0) + b0)
```

```python
t = tf.placeholder(tf.float32, [None, 1])
costFunction = -tf.reduce_sum(t*tf.log(p) + (1-t)*tf.log(1-p))
trainOptimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0017)
train = trainOptimizer.minimize(costFunction)
prediction = tf.equal(tf.sign(p-0.5), tf.sign(t-0.5))
accuracy = tf.reduce_mean(tf.cast(prediction, tf.float32))
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```

```python
itrNum = 0
for _ in range(2000):
    itrNum += 1
    sess.run(train, feed_dict={x:train_x, t:train_t})
    if itrNum % 100 == 0:
        costFunctionValue, accuracyValue = sess.run([costFunction, accuracy], feed_dict={x:train_x, t:train_t})
        print ('Step: %d, Loss: %f, Accuracy: %f' % (itrNum, costFunctionValue, accuracyValue))
```
